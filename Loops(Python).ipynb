```python
"Finding sum of every list element using For loop"
li = [10,30,5,33,52,4,443,5]

sum = 0
for i in li:
    sum += i
print('Sum is = ', sum)
```

```python
"Finding sum of every list element using While loop"
li = [192,4,15,3,34]
sum = 0
i = 0
while i<len(li):
    sum += li[i]
    i += 1
print('Sum is = ', sum)
```

```python
"Printing a right angle triange "
for i in range(1,11):
    for j in range(11-i-1):
        print(" ", end="")
    for j in range(i):
        print('*', end="")
    print()
```

```python
"Printing a equilateral triange"
n = 40
for i in range(1,11):
    print("  "*n, end=" ")
    print(' * '*(i))
    n -= 1
```

```python
"use of break statement"
x = 10
while x >= 1:
    print('x = ', x)
    x -= 1
    if x == 5:
        break
print('Out of loop because of break')
```

```python
"The continue statement"
x = 0
while x <= 10:
    x += 1
    if x > 5:
        continue
    print('x = ', x)
print('Out of loop')
```

```python
"The pass statement"

x = 0
while x < 10:
    x += 1
    if x > 5:
        pass
    print('x = ', x)
print('Out of the loop')
```

```python
"Prime number print without function"

ax  = int(input("Upto what number? "))
for num in range(2, ax+1):
    for i in range(2, num):
        if (num % i ) == 0:
            break
        else:
            print(num)
```

```python
"Fibbonacci series without function"

n = int(input('Enter how many fibbonacci? '))
f1 = 0
f2 = 1
c = 2

if n == 1:
    print(f1)
elif n == 2:
    print(f1, '\n', f2, sep='')
else:
    print(f1, '\n', f2, sep='')
    while c < n:
        f = f1 + f2
        print(f)
        f1, f2 = f2, f
        c += 1
```
